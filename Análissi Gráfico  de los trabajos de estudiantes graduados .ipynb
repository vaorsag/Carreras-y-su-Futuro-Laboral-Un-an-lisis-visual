# Carreras y su Futuro Laboral: Un análisis visual
Este trabajo consiste en el análisis de los estudiantes graduados de la universidad (generación 2010-2012) y su actual situación económica. Se explorará cuestiones como:
 * ¿ En que carreras se gana mayor sueldo?
 * ¿ Cuales carreras tienen mas estudiantes?
 * ¿ Como se distribuye la preferencia de carreras segun sexo?
 
Los datos utilizada son de American Community Survey y el dataset junto con la descripción de los encabezados puede ser encontrado [Aqui](https://github.com/fivethirtyeight/data/tree/master/college-majors)

## A . Data Cleaning 

Se comienza el trabajo haciendo una importación de las librerias necesarias, la lectura del archivo a analizar y la limpieza de las filas que contienen valores nulos.

### Exploración de los datos

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv') #dataframe
print("Encabezado:\n",recent_grads.iloc[0])  #encabezado
print("\n Primeras filas:\n",recent_grads.head(2)) # explorando los primeros datos
print("\n Ultimas filas:\n",recent_grads.tail(2)) #explorando los últimos datos
print("\n Resumen estadístico:\n")
recent_grads.describe() # estadistica de los datos
```

### Limpieza de filas

Puesto que se utilizarán las columna de datos para realizar gráficos, es necesario eliminar los datos inexistentes, puesto que el uso de matplotlib con valores inexistentes arrojaria error. 

Se realiza un conteo de cuantas filas se eliminaron para asegurarse que no fueron demasiadas. 

```python
raw_data_count = recent_grads.shape[0] #saber el número de filas
print(raw_data_count)
```

```python
recent_grads.dropna(axis = 0, inplace=True) #eliminar filas con nulls
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

Se observa que solamente una fila fue eliminada
